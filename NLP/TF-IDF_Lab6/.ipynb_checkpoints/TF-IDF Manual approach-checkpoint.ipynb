Take a para. Remove stopwords. Create unique words list. Among the unique words if a word is synonymous, then remove it
and keep only one word.
Then find the, frequency of each word. Find the Term Frequency (TF) and find how many sentences it repeats in.

Then find IDF. IDF = log(DF/N) where N is Total number of sentences. 

Then find weight of each word using TF * IDF

## Reading the text document

```python
import pandas as pd
import re
from nltk.corpus import stopwords, wordnet
import nltk
from collections import Counter
import math
```

```python
with open('text.txt', 'r') as f:
    lines = f.readlines()
```

```python
sentences = lines
sentences = " ".join(sentences)

```

## Preprocessing the text


```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s.]', '', text)
    text = re.sub(r'\b(\w+)\.', r'\1', text)
    
    return text

preprocessed_paragraph = preprocess_text(sentences)
```

## Removing stopwords

```python
def remove_stopwords(text, stopwords):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stopwords]
    return " ".join(filtered_words)
```

```python
text_without_stopwords = remove_stopwords(preprocessed_paragraph, stop_words)
```

```python
words = text_without_stopwords.split()
```

```python
def get_synonyms(word):
    synonyms = set()
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
    return synonyms

def remove_synonymous_words(words):
    unique_words = set(words)
    words_to_remove = set()
    for word in unique_words:
        synonyms = get_synonyms(word)
        for syn in synonyms:
            if syn in unique_words:
                words_to_remove.add(syn)
    
    unique_words.difference_update(words_to_remove)
    return list(unique_words)
```

```python
unique_words = remove_synonymous_words(words)
```

```python
def calculate_tf(text, unique_words):
    words = text.split()
    word_count = Counter(words)
    tf = {word: word_count.get(word, 0) for word in unique_words}
    return tf

# Example usage:
tf = calculate_tf(" ".join(words), unique_words)
tf
```

```python
def calculate_df(corpus):
    df = Counter()
    for document in corpus:
        words = set(document.split())
        df.update(words)
    return df

df = calculate_df(lines)
```

```python
def calculate_idf(df, N):
    idf = {word: math.log(N / df[word]) for word in df}
    return idf

N = len(lines)
idf = calculate_idf(df, N)
```

```python
def calculate_tfidf(tf, idf):
    tfidf = {word: tf[word] * idf.get(word, 0) for word in tf}
    return tfidf

tfidf = calculate_tfidf(tf, idf)
```

```python
print("TF-IDF for each word:")
for word, score in tfidf.items():
        print(f"{word}: {score}")
```
