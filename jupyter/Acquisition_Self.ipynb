```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
```

```python
url = 'https://en.wikipedia.org/wiki/Public_holidays_in_India'
data = urlopen(url)
```

```python
fest_html = data.read()
print(fest_html)
```

```python
data_soup = soup(fest_html, 'html.parser')
print(data_soup.prettify)
```

```python
data_soup.find('table')
```

```python
tables = data_soup.find_all('table',{'class': 'wikitable'})
```

```python
data_table = tables[1]
print(data_table)
```

```python
data_header = data_table.find_all('th')
print(data_header)
```

```python
coloumn_headers = [i.text[:-1] for i in data_header]
print(coloumn_headers)
```

```python
rows_data = data_table.find_all('tr')[1:]
first = rows_data[0]
first
```

```python
first = rows_data[0].find_all('td',())
for d in first:
    print(d.text)
```

```python
table_rows = []

for row in rows_data:
    curr_row = []
    row_data = row.find_all('td', {})
    for idx,data in enumerate(row_data):
        if idx == 0:
            s = data.text
            s = s.replace('\n','')
            s = s.replace(',','')
            curr_row.append(s)
        elif idx == 1:
            s = data.text
            s = s.replace('\n','')
            s = s.replace(',','')
            curr_row.append(s)
        elif idx == 2:
            s = data.text[:-1]
            s = s.replace(',','')
            s = s.replace('\n','')
            curr_row.append(s)
        else:
            curr_row.append(data.text)
        
    table_rows.append(curr_row)

table_rows
```

```python
file_name = 'fest_india.csv'

with open(file_name, 'w') as f:
    # Write the header first
    
    header_string = ','.join(coloumn_headers)
    header_string += '\n'
    f.write(header_string)
    
    for row in table_rows:
        row_string = ""
        
        for w in row:
            row_string += w + ','
        
        row_string = row_string[:-1]
        row_string += '\n'
        f.write(row_string)
```
