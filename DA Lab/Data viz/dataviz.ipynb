```python
x = np.array([1, 2, 3, 4])

plt.plot(x, x + 1, linestyle = 'solid')
plt.plot(x, x + 1.5, linestyle = 'dashed')
plt.plot(x, x + 2, linestyle = 'dashdot')
plt.xlabel('Value of X')
plt.ylabel('Value of Y')
plt.title('Line plot')

plt.show()
```
