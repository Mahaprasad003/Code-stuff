```python
x=np.array([1,2,3,4])
```

```python
plt.plot(x,x+1, linestyle = 'solid')
plt.plot(x,x+1.5,linestyle = 'dashed')
plt.plot(x ,x+2, linestyle = "dashdot")
plt.xlabel("value of x")
plt.ylabel("value of incremented by .5")
plt.title("line plot")
plt.show()
```
